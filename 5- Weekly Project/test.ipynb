```python
import pandas as pd 

df = pd.read_csv('D:\\Tuwaiq Academy\\Tuwaiq_Academy_T5_Week_6\\5- Weekly Project\\Datasets\\Traffic Data.csv')
df
```

```python
df.set_index('date_time', inplace=True)
df.sort_index()
df
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder()

df[['weather_description', 'weather_main']] = oe.fit_transform(df[['weather_description', 'weather_main']])
df
```

```python
len(df)
```

```python
from sklearn.model_selection import train_test_split
train_val , test = train_test_split(df, shuffle=False, test_size=0.2)
train , val = train_test_split(df, shuffle=False, test_size=0.2)
train.shape, test.shape, val.shape
```

```python
train_traget = train[['traffic_volume', 'temp']].iloc[6:]
test_traget = test[['traffic_volume', 'temp']].iloc[6:]
val_traget = val[['traffic_volume', 'temp']].iloc[6:]
```

```python
from keras.utils import timeseries_dataset_from_array

# timeseries_dataset_from_array?
Xy_train = timeseries_dataset_from_array(train, targets=train_traget, sequence_length=6, shuffle=False)
Xy_test = timeseries_dataset_from_array(test, targets=test_traget, sequence_length=6, shuffle=False)
Xy_val = timeseries_dataset_from_array(val, targets=val_traget, sequence_length=6, shuffle=False)
```

```python
for x, y in Xy_train:
    print(x, y)
```

```python
X_train, y_train = next(iter(Xy_train))
X_test, y_test = next(iter(Xy_test))
X_val, y_val = next(iter(Xy_val))
```

```python
X_train.shape
```

```python
Xy_train
```

```python
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Dropout
window_len = 6

model = Sequential([
    SimpleRNN(50, return_sequences=True, input_shape=(window_len, 7), activation='tanh'),
    Dropout(0.5),
    SimpleRNN(50),
    Dropout(0.5),
    Dense(2, activation='relu')
])

model.compile(loss='mean_squared_error', optimizer='adam', metrics=['mae'])

model.summary()
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
checkpoints = ModelCheckpoint(filepath = 'my_weights.keras', save_best_only = True)

early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True)

model.fit(X_train, y_train,
          validation_data=(X_val, y_val),
          epochs=100,
          batch_size=32,
          verbose=1,
          callbacks= [checkpoints, early_stopping])
```

```python
# Let's do the prediction and check performance metrics
train_predict=model.predict(X_val)
```

```python
from sklearn.metrics import mean_squared_error
import math
print(math.sqrt(mean_squared_error(y_val, train_predict)))
```
